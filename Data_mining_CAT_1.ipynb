```python
import pandas as pd
from sklearn.cluster import KMeans

df=pd.read_csv("cat1.csv")
df
```

```python
unique_items = list(set(item for sublist in df['Items_Purchased'].str.split(', ') for item in sublist))
print(unique_items)

oht = pd.DataFrame(columns=unique_items)

for index, row in df.iterrows():
    itemset = row['Items_Purchased'].split(', ')
    oht.loc[index, itemset] = 1

oht.fillna(0, inplace=True)
print(oht)

kmeans = KMeans(n_clusters=3)
kmeans.fit(oht)

oht['Cluster'] = kmeans.labels_
print(oht)
```

```python
df['Cluster']= oht['Cluster']
df
```

```python
c=list(df['Cluster'])
print(c)

t=list(df['Items_Purchased'])
tl=[]
for i in t:
  l=list(i.split(', '))
  tl.append(l)

print(tl)
```
